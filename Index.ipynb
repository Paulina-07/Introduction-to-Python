# INTRODUCTION TO PYTHON

The history of Python:
Python is a programming language created by the Dutch Guido van Rossum in December 1989, giving continuity to the ABC programming language.
In February 1991, Van Rossum published the code for version 0.9.0 in which exception handling, functions and the modular types ready, dict, ser and so on were already present. Van Rossum described the module as one of the major Python programming units, Python's exception model is similar to Modula-3 but with the addition of an else capsule.

In January 1994 version 1.0 of Python came out with new programming tools: lambda, map, filter and reduce.
On October 16, 2000 Python version 2.0 was released with new features such as full garbage collection and full Unicode support. Python 2.0 was the first and only release of BeOpen.con. After the launch Guido van Rossum along with the other developers teamed up with Digital Creations. As of version 2.0 Python took a major feature of the functional programming language Haskell: List comprehensions. Python's syntax is very similar to Haskell's, the difference between the two being Haskell's preference for punctuation characters and Python's preference for alphabetic keywords. Version 2.0 also included a garbage collection system capable of collecting cyclic references, list comprehensions, augmented assignments, and new string methods.

The Python 3.0 version was released on December 3, 2008 and among the main changes are the separation of Unicode strings and binary data, the print function, changes in syntax, data types, comparators, etc.
Currently Python is the third most used programming language in the world.
